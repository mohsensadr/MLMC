```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("mlmc_vs_mc.csv")

plt.figure(figsize=(3, 3))
plt.loglog(df["eps"], df["mlmc_cost"], "o-", label="MLMC")
plt.loglog(df["eps"], df["mc_cost"], "s-", label="MC")
plt.xlabel("ε")
plt.ylabel("Cost")
plt.legend()
plt.grid(True, which="both")
plt.savefig('mlmc_vs_mc_cost.png', bbox_inches='tight')

plt.figure(figsize=(3, 3))
df = pd.read_csv("mlmc_details.csv")
eps = df["eps"]
L = df["L"]
Nl = df["Nl"]
for i in range(L.shape[0]):
    # convert string list to integers
    nl_values = list(map(int, Nl[i].split()))
    plt.plot(range(len(nl_values)), nl_values, marker="o", label=f"ε={eps[i]}")
plt.xlabel(r"no. levels")
plt.legend()
plt.yscale("log")
plt.ylabel("no. of samples per level")
plt.grid(True)
plt.savefig('mlmc_Nl_eps.png', bbox_inches='tight')
plt.show()
```
